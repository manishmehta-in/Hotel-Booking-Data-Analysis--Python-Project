# Importing Libraries

```python
import pandas as pds
import matplotlib.pyplot as mpl
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pds.read_csv("hotel_bookings.csv")
```

# Data Analysis and Visualizations

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)

mpl.figure(figsize = (5,4))
mpl.title('Reservation status count')
mpl.bar(['Not Canceled', 'Canceled'],df['is_canceled'].value_counts(),edgecolor = 'k', width = 0.7)
mpl.show()
```

```python
mpl.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))
mpl.title('Reservation status in different hotels', size = 20)
mpl.xlabel('hotel')
mpl.ylabel('number of reservations')
mpl.legend(['not canceled','canceled'])
mpl.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
mpl.figure(figsize = (20,8))
mpl.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
mpl.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
mpl.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
mpl.legend(fontsize = 20)
mpl.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
mpl.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'bright')
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
mpl.title('Reservation status per month', size = 20)
mpl.xlabel('month')
mpl.ylabel('number of reservations')
mpl.legend(['not canceled', 'canceled'])
mpl.show()
```

```python
mpl.figure(figsize=(15, 8))
mpl.title('ADR per month', fontsize=30)
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
mpl.show()
```

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
mpl.figure(figsize = (8,8))
mpl.title('Top 10 countries with reservation canceled')
mpl.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
mpl.show()
```

```python
canceled_data['market_segment'].value_counts(normalize = True)
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

mpl.figure(figsize = (20,6))
mpl.title('Average Daily Rate')
mpl.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label = 'not cancelled')
mpl.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'cancelled')
mpl.legend()
```

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
mpl.figure(figsize = (20,6))
mpl.title('Average Daily Rate', fontsize = 30)
mpl.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'not cancelled')
mpl.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label = 'cancelled')
mpl.legend(fontsize = 20)
mpl.show()
```
